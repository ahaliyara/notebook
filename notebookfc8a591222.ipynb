```python
import pandas as pd  # For data manipulation
import numpy as np  # For numerical operations
import seaborn as sns  # For visualization
import matplotlib.pyplot as plt  # For plotting
```

```python
df = pd.read_csv("/kaggle/input/test-file/tested.csv")  # Load the dataset
df.head()  # Show first 5 rows
```

```python
df.isnull().sum()
```

```python
df['Age'].fillna(df['Age'].median(), inplace=False)
```

```python
df.drop(columns=['Cabin'], inplace=False)
```

```python
sns.barplot(x='Sex', y='Survived', data=df)
plt.title("Survival Rate by Gender")
```

```python
sns.barplot(x='Pclass', y='Survived', data=df)
plt.title("Survival Rate by Passenger Class")
```

```python
plt.figure(figsize=(10,5))
sns.histplot(df[df['Survived']==1]['Age'], bins=30, kde=True, color="green", label="Survived")
sns.histplot(df[df['Survived']==0]['Age'], bins=30, kde=True, color="red", label="Not Survived")
plt.legend()
plt.title("Age Distribution of Survivors vs. Non-Survivors")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set Seaborn style for better aesthetics
sns.set_style("whitegrid")

# Create the figure
plt.figure(figsize=(10, 5), dpi=100)

# Plot histograms with improved styling
sns.histplot(df[df['Survived'] == 1]['Age'], bins=30, kde=True, color="green", label="Survived", alpha=0.6)
sns.histplot(df[df['Survived'] == 0]['Age'], bins=30, kde=True, color="red", label="Not Survived", alpha=0.6)

# Add title and labels
plt.title("Age Distribution of Survivors vs. Non-Survivors", fontsize=14, fontweight="bold")
plt.xlabel("Age", fontsize=12)
plt.ylabel("Count", fontsize=12)

# Add a legend
plt.legend(title="Survival Status", fontsize=11)

# Display the plot
plt.show()
```

```python
df['FamilySize'] = df['SibSp'] + df['Parch'] + 1
sns.barplot(x='FamilySize', y='Survived', data=df)
plt.title("Survival Rate by Family Size")
```

```python
plt.figure(figsize=(10,5))
sns.boxplot(x='Pclass', y='Fare', data=df)
plt.title("Fare Distribution by Class")
```
